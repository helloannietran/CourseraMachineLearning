# Logistic Regression (OnevsAll)

```python
def apply_sigmoid(value):
    
    result = 1/(1 + np.exp(-value))
    
    return result


def compute_reg_cost(theta, x, y, lmbda):
    
    m = len(y)
    temp1 = np.multiply(y, np.log(apply_sigmoid(np.dot(x, theta))))
    temp2 = np.multiply(1-y, np.log(1-apply_sigmoid(np.dot(x, theta))))
    
    cost = np.sum(temp1 + temp2) / (-m) + np.sum(theta[1:]**2) * lmbda / (2*m)

    return cost
    

def reg_gradient_descent(theta, x, y, lmbda):
    
    m = len(y)
    temp = apply_sigmoid(np.dot(x, theta)).reshape(y.shape) - y
    temp = np.dot(temp.T, x).T / m + theta * lmbda / m
    temp[0] = temp[0] - theta[0] * lmbda / m
    
    return temp
```

```python
k = 10 # ten classes
theta = np.zeros((k, X.shape[1]))
lambd = 0.1
```

```python
for k in range(k):
    print(f'Fitting model for digit {k}')
    i = k if k else 10

    theta[k] = fmin_cg(f=compute_reg_cost, x0=theta[k],  fprime=reg_gradient_descent,
                       args=(X, (y == i).flatten(), lambd), maxiter=50)
```

```python
theta.shape
```

```python
# Making prediction and score accuracy

pred = np.argmax(np.dot(X,theta.T), axis = 1)
pred = [e if e else 10 for e in pred]
np.mean(pred == y.flatten()) * 100
```

# Neural Network

```python
# load pretrained weights
weights = loadmat('ex3weights.mat')
theta1 = weights['Theta1']
theta2 = weights['Theta2']

# swap first and last columns of Theta2, due to legacy from MATLAB indexing, 
# since the weight file ex3weights.mat was saved based on MATLAB indexing
theta2 = np.roll(theta2, 1, axis=0)
```

```python
print(theta1.shape)
print(theta2.shape)
```

```python
# add in bias
new_X = np.hstack((np.ones((X.shape[0], 1)), X))
new_X.shape
```

```python
nn_preds_1 = apply_sigmoid(np.dot(new_X, theta1.T))
nn_preds_1 = np.hstack((np.ones((nn_preds_1.shape[0],1)), nn_preds_1))

nn_preds_2 = np.argmax(apply_sigmoid(np.dot(nn_preds_1, theta2.T)), axis=1)
nn_preds_2 = [v if v else 10 for v in nn_preds_2]
```

```python
np.mean(nn_preds_2 == y.flatten()) * 100
```
